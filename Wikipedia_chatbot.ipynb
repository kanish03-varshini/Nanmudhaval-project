```python

# STEP 1: Install required libraries
!pip install wikipedia langchain google-generativeai langchain-google-genai

# STEP 2: Imports
import wikipedia
import os
from langchain.prompts import PromptTemplate
from langchain_google_genai import ChatGoogleGenerativeAI

# STEP 3: Set your Gemini API key
os.environ["GOOGLE_API_KEY"] = "AIzaSyBNbADhbEvFQHKV4SogbBK1zW8oQfwSsRE"  # <-- Replace with your actual Gemini API key

# STEP 4: Initialize Gemini model
llm = ChatGoogleGenerativeAI(
    model="models/gemini-1.5-pro-latest",
    temperature=0.5,
    google_api_key=os.environ["GOOGLE_API_KEY"]
)

# STEP 5: Define prompt template
template = """
You are a helpful AI assistant. Based on the Wikipedia information below, answer the user's question.

Wikipedia Content:
{context}

User Question:
{question}
"""

prompt = PromptTemplate.from_template(template)

# STEP 6: Search Wikipedia and get summary
def get_wikipedia_summary(topic):
    try:
        return wikipedia.summary(topic, sentences=5)
    except Exception as e:
        return f"Error: {str(e)}"

# STEP 7: Chat loop
print("Wikipedia Chatbot Ready! Type 'exit' to stop.")
while True:
    user_query = input("You: ")
    if user_query.lower() == "exit":
        print("Chat ended.")
        break

    wiki_content = get_wikipedia_summary(user_query)
    chain = prompt | llm
    response = chain.invoke({"context": wiki_content, "question": user_query})
    print("Bot:", response.content)
```
